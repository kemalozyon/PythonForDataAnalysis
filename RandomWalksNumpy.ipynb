# Random Walk with one dimensional Array

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_steps = 1000
start_pos = 0
```

```python
steps = np.random.choice([-1,1], size=n_steps)
```

```python
walk = np.cumsum(steps)
```

```python
walk_with_start = np.insert(walk, 0, start_pos)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(walk_with_start)
plt.title("1D Random Walk with Python")
plt.xlabel("Step Count")
plt.ylabel("Location")
plt.grid(True)
plt.show()
```

# Random Walk with 2 Dimensional Array

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_steps = 5000
start_pos =[0, 0]
```

```python
steps_x = np.random.choice([-1, 1], size=n_steps)
steps_y = np.random.choice([-1, 1], size=n_steps)
```

```python
walk_x = np.cumsum(steps_x)
walk_y = np.cumsum(steps_y)
```

```python
path_x = np.insert(walk_x, 0, start_pos[0])
path_y = np.insert(walk_y, 0, start_pos[1])
```

```python
plt.figure(figsize=(10,10))
plt.plot(path_x, path_y, 'o-', markersize=2, alpha=0.7)
plt.plot(path_x[0], path_y[0], 'ro', markersize=5)
plt.plot(path_x[-1], path_y[-1], 'go', markersize=5)
plt.title("2D Random Walk with Python")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.axis("equal")
plt.show()
```
